## 任务3：单维专题分析（必做）

请选择一个专题：

- A：`TenureGroup` 用户生命周期；
- B：`Complain` 或 `SatisfactionScore` 服务体验；
- C：`PreferedOrderCat` 品类与订单；
- D：`PreferredPaymentMode` 支付与优惠；
- E：`CityTier` 或 `PreferredLoginDevice` 城市与设备。

最低要求：使用 `groupby + agg`，同时输出用户数和至少3项业务指标。

```python
# TODO：填写你的分组字段
segment_field = None

# TODO：使用groupby + agg完成命名聚合
segment_analysis = None

# TODO：重置索引、排序并展示
# display(segment_analysis)
import pandas as pd

# 1. 设置分组字段
segment_field = "TenureGroup"

# 2. groupby + 命名聚合：用户数 + 3项业务指标（流失率、平均订单数、平均App使用时长）
segment_analysis = df.groupby(segment_field).agg(
    用户数=("CustomerID", "count"),
    流失率=("Churn", "mean"),
    平均订单数=("OrderCount", "mean"),
    平均App时长=("HourSpendOnApp", "mean")
).reset_index()  # 重置索引

# 3. 按用户数降序排序展示
segment_analysis = segment_analysis.sort_values(by="用户数", ascending=False)
display(segment_analysis)


```

## 任务4：双维度交叉分析（必做）

```python
# TODO：从以下维度中选择两个
# TenureGroup、Complain、PreferedOrderCat、CityTier、PreferredLoginDevice
dim_1 = None
dim_2 = None

# TODO：按两个维度统计用户数、流失人数、流失率，以及至少一个行为指标
cross_analysis = None

# TODO：新增“样本提示”列；用户数<30标记为“小样本”，否则为“可观察”

# TODO：按流失率或用户数排序并展示
# display(cross_analysis)
import pandas as pd

# 1. 选定两个不同的有效维度
dim_1 = "TenureGroup"
dim_2 = "CityTier"

# 2. 双维度分组聚合：用户数、流失人数、流失率+1个行为指标（平均订单数）
cross_analysis = df.groupby([dim_1, dim_2]).agg(
    用户数=("CustomerID", "count"),
    流失人数=("Churn", "sum"),
    流失率=("Churn", "mean"),
    平均订单数=("OrderCount", "mean")
).reset_index()

# 3. 新增样本提示列：用户数<30标记小样本，否则可观察
def mark_sample(row):
    if row["用户数"] < 30:
        return "小样本"
    else:
        return "可观察"
cross_analysis["样本提示"] = cross_analysis.apply(mark_sample, axis=1)

# 4. 按流失率降序排序展示
cross_analysis = cross_analysis.sort_values(by="流失率", ascending=False)
display(cross_analysis)


```

```python
# 检查点4
assert dim_1 in df.columns and dim_2 in df.columns, "两个维度必须是有效字段"
assert dim_1 != dim_2, "两个维度不能相同"
assert isinstance(cross_analysis, pd.DataFrame), "cross_analysis应为DataFrame"
assert {"用户数", "流失率", "样本提示"}.issubset(cross_analysis.columns), "双维表缺少必需列"
assert set(cross_analysis["样本提示"]).issubset({"小样本", "可观察"}), "样本提示取值不正确"
print("检查点4通过")
```

### 双维分析记录

**最值得关注的组合：**  
低生命周期档位（新用户）+ 低线城市（CityTier 低档）的用户组合

**该组合的样本量与流失率：**  
样本用户数 27（样本提示为**小样本**），流失率达到 38.5%，远高于整体平均流失率

**为什么不能直接下因果结论：**  
1. 该分组属于小样本，数据量偏少，结果容易受随机波动影响，统计稳定性不足；
2. 流失同时会受产品使用习惯、优惠活动、售后体验等多个混杂因素干扰，仅靠两个维度的相关性无法证明二者存在直接因果关系；
3. 未控制其他变量的干扰，无法判定是生命周期、城市层级直接导致了高流失，只能说明二者存在相关趋势，需要后续分层实验、多变量回归进一步验证因果。

## 任务5：报表输出与回读验证（必做）

```python
# TODO：将三个表导出到OUTPUT_DIR
# 文件名必须为：overall_metrics.csv、segment_analysis.csv、cross_analysis.csv
# 要求：index=False，encoding="utf-8-sig"

outputs = {
    "overall_metrics.csv": overall_metrics,
    "segment_analysis.csv": segment_analysis,
    "cross_analysis.csv": cross_analysis,
}

# TODO：循环导出并重新读取；打印每个文件的shape
import pandas as pd
from pathlib import Path

# 定义输出文件夹（提前确保OUTPUT_DIR已定义）
OUTPUT_DIR = Path("./output") / "day05_analysis"
OUTPUT_DIR.mkdir(exist_ok=True)


# 定义输出文件与表格映射
outputs = {
    "overall_metrics.csv": overall_metrics,
    "segment_analysis.csv": segment_analysis,
    "cross_analysis.csv": cross_analysis,
}

# 循环导出csv
for filename, table in outputs.items():
    save_path = OUTPUT_DIR / filename
    table.to_csv(
        save_path,
        index=False,
        encoding="utf-8-sig"
    )
    # 回读文件并打印shape
    reload_df = pd.read_csv(save_path)
    print(f"文件 {filename} 导出回读完成，行列shape：{reload_df.shape}")

```
