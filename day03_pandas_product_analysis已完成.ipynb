## 任务4：条件筛选与排序

```python
# TODO 1：筛选广东商品，变量名为guangdong
guangdong = None

# TODO 2：筛选广东且商品价格不低于1000元的商品
# 只保留商品id、一级品类、二级品类、商品价格、省份、商品销量
# 按商品价格从高到低排序，变量名为guangdong_high_price
guangdong_high_price = None

# TODO 3：筛选浙江或江苏商品，变量名为zhejiang_or_jiangsu
zhejiang_or_jiangsu = None

# TODO 4：展示广东高价商品前10行和浙江或江苏商品数
import pandas as pd

# TODO 1：筛选广东商品
guangdong = df[df["省份"] == "广东"]

# TODO 2：筛选广东且价格≥1000元，指定列并按价格降序排序
cols_need = ["商品id", "一级品类", "二级品类", "商品价格", "省份", "商品销量"]
guangdong_high_price = df[(df["省份"] == "广东") & (df["商品价格"] >= 1000)][cols_need]
guangdong_high_price = guangdong_high_price.sort_values(by="商品价格", ascending=False).reset_index(drop=True)

# TODO 3：筛选浙江或江苏商品
zhejiang_or_jiangsu = df[df["省份"].isin(["浙江", "江苏"])]

# TODO 4：展示结果
print("广东高价商品前10行：")
display(guangdong_high_price.head(10))
print("\n浙江或江苏商品总数量：", len(zhejiang_or_jiangsu))
```

## 任务5：描述性统计与一级品类汇总

```python
# TODO 1：使用describe查看商品价格摘要，变量名为price_summary
price_summary = None

# TODO 2：按一级品类统计商品数、平均价格和中位价格
# 按平均价格从高到低排序，变量名为category_summary
category_summary = None

# TODO 3：展示结果
import pandas as pd

# TODO 1：使用describe查看商品价格摘要
price_summary = df["商品价格"].describe()

# TODO 2：按一级品类聚合统计，再排序
category_summary = df.groupby("一级品类").agg(
    商品数=("商品id", "count"),
    平均价格=("商品价格", "mean"),
    中位价格=("商品价格", "median")
).sort_values(by="平均价格", ascending=False).reset_index()

# TODO 3：展示结果
print("商品价格整体描述统计：")
display(price_summary)
print("\n一级品类汇总统计（按均价降序）：")
display(category_summary)
```

```python
# 检查点5
assert isinstance(price_summary, pd.Series)
assert isinstance(category_summary, pd.DataFrame)
assert {"商品数", "平均价格", "中位价格"}.issubset(category_summary.columns)
assert category_summary["商品数"].sum() == len(df)
assert category_summary["平均价格"].is_monotonic_decreasing
assert abs(df["商品价格"].mean() - 938.26) < 0.02
print("检查点5通过")
```

## 挑战任务：省份—类别分析

```python
# TODO 1：选择两个省份
provinces = ["广东", "江苏"]

# TODO 2：筛选省份并统计商品数、平均价格和中位价格
province_summary = None

# TODO 3：分别计算两个省份最常见的一级品类
top_categories = None

# TODO 4：展示结果
import pandas as pd

# TODO 1：选定两个省份
provinces = ["广东", "江苏"]

# TODO 2：筛选省份，按省份分组统计指标
filter_df = df[df["省份"].isin(provinces)]
province_summary = filter_df.groupby("省份").agg(
    商品数=("商品id", "count"),
    平均价格=("商品价格", "mean"),
    中位价格=("商品价格", "median")
)

# TODO 3：分别统计两省出现频次最高的一级品类，整合为DataFrame
gd_top = filter_df[filter_df["省份"] == "广东"]["一级品类"].value_counts().idxmax()
js_top = filter_df[filter_df["省份"] == "江苏"]["一级品类"].value_counts().idxmax()
top_categories = pd.DataFrame({
    "省份": provinces,
    "最常见一级品类": [gd_top, js_top]
}).set_index("省份")

# TODO 4：展示结果
print("省份汇总统计：")
display(province_summary)
print("\n两省最高频一级品类：")
display(top_categories)
```

```python
# 检查点6
assert len(provinces) == 2 and provinces[0] != provinces[1]
assert isinstance(province_summary, pd.DataFrame)
assert set(province_summary.index) == set(provinces)
assert {"商品数", "平均价格", "中位价格"}.issubset(province_summary.columns)
assert isinstance(top_categories, pd.DataFrame)
print("检查点6通过")
```
